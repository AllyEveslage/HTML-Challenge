## Deliverable 2: Scrape and Analyze Mars Weather Data

```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
browser = Browser('firefox')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
mars_soup = soup(html, "html.parser")
mars_soup
```

```python
# Extract all rows of data
table_elements = mars_soup.find_all("tr", class_="data-row")
table_elements
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Heres an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
new_list = []

# Loop through the scraped data to create a list of rows
for row in table_elements:
    cell_values = row.find_all("td")
    values = []
    for value in cell_values:
        values.append(value.text)
    new_list.append(values)

new_list
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
temp_df = pd.DataFrame(new_list, columns = ['id', 'terrestrial_date', 'sol', 'ls', 'month', 'min_temp', 'pressure'])
```

```python
# Confirm DataFrame was created successfully
temp_df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
temp_df.dtypes
```

```python
# Change data types for data analysis
temp_df["id"]= pd.to_numeric(temp_df["id"])
temp_df["terrestrial_date"] = pd.to_datetime(temp_df["terrestrial_date"])
temp_df["sol"] = pd.to_numeric(temp_df["sol"])
temp_df["ls"] = pd.to_numeric(temp_df["ls"])
temp_df["month"] = pd.to_numeric(temp_df["month"])
temp_df["min_temp"] = pd.to_numeric(temp_df["min_temp"])
temp_df["pressure"] = pd.to_numeric(temp_df["pressure"])
```

```python
# Confirm type changes were successful by examining data types again
temp_df.dtypes
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
temp_df['month'].max()
```

```python
# 2. How many sols (Martian days) worth of data are there?
temp_df['sol'].count()
```

```python
# 3. What is the average minimum temperature by month?
avg_min_temp = temp_df["min_temp"].groupby(temp_df['month']).mean()
avg_min_temp
```

```python
# Plot the average minimum temperature by month
avg_min_temp.plot.bar(rot = 0, color = 'mediumpurple', figsize=(10,4), fontsize = 10, alpha = 0.55)
plt.ylabel("Average Low Temp")
plt.xlabel("Month")
plt.title("Average Lowest Temp per Month on Mars", fontsize = 12)
plt.grid(axis = 'y')
```

```python
# Identify the coldest and hottest months in Curiosity's location by sorting the previous graph
# Sorting the data by temp in ascending order (coldest to hottest)
avg_min_temp_sorted = avg_min_temp.sort_values()

avg_min_temp_sorted.plot.bar(rot=0, color='mediumpurple', figsize=(10,4), fontsize=10, alpha=0.55)
plt.ylabel("Average Low Temp")
plt.xlabel("Month")
plt.title("Average Lowest Temp per Month on Mars (Sorted)", fontsize=12)
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
# 4. What is the average pressure by month?
avg_pressure_monthly = temp_df["pressure"].groupby(temp_df["month"]).mean()
avg_pressure_monthly
```

```python
# Plot the average pressure by month
avg_pressure_monthly.plot.bar(rot = 0, color = 'mediumpurple', figsize=(10,4), fontsize = 10, alpha = 0.55)
plt.ylabel("Pressure (mmHg)")
plt.xlabel("Month")
plt.title("Average Pressure (per mmHg) per Month on Mars", fontsize = 12)
plt.grid(axis = 'y')
```

```python
# Identify the lowest and highest pressure months in Curiosity's location by sorting the previous graph
avg_pressure_sorted = avg_pressure_monthly.sort_values()

avg_pressure_sorted.plot.bar(rot = 0, color = 'mediumpurple', figsize=(10,4), fontsize = 10, alpha = 0.55)
plt.ylabel("Pressure (mmHg)")
plt.xlabel("Month")
plt.title("Average Pressure (per mmHg) per Month on Mars", fontsize = 12)
plt.grid(axis = 'y')
```

```python
# 5. How many terrestrial (Earth) days are there in a Martian year?
mars_year_days = (datetime.strptime('2015-06-17', "%Y-%m-%d") - 
                  datetime.strptime('2013-08-01', "%Y-%m-%d")).days

print(mars_year_days)
```

```python
avg_min_temp_per_month = mars_df.groupby('sol')['min_temp'].mean()

# Plot the data
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(avg_min_temp_per_month.index, avg_min_temp_per_month.values, linestyle='-', color = 'mediumpurple', alpha = 0.55)

# Labels and title
ax.set_title("Average Minimum Daily Temperature per Month on Mars", fontsize=15)
ax.set_xlabel("Number of Terrestrial Days", fontsize=12)
ax.set_ylabel("Min Temperature (F)", fontsize=12)

# Show the plot
plt.show()
```

#### Minimum Temperature

When observing the graphs above which follow the minimum temp per month, we can see that, on average, the third month on Mars is the coldest month. On the contrary, the eighth month, on average, tends to be the warmest. As for us humans, no matter what month, Mars sure is cold!

#### Atmospheric Pressure

When observing the graphs above which follow the average atmospheric pressure per month, we can see that, on average, the sixth month is the lowest in pressure, while the ninth month is the highest in pressure. 

#### Year Length

Using the above observations, we're able to see a year on Mars appears to be 685 days (45 days short of DOUBLE what one year on Earth is!). We can cross-check this information by looking at how far Mars is from the sun in comparison to that of Earth.
